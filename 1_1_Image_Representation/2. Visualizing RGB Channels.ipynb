# RGB colorspace

### Import resources

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

### Read in an image

```python
# Read in the image
image = mpimg.imread('images/wa_state_highway.jpg')

plt.imshow(image)
```

```python
print(f'Image Shape: {image.shape}')
print(f'Min Color Intensity: {np.amin(image)}')
print(f'Max Color Intensity: {np.amax(image)}')
print(f'Image Data Type: {image.dtype}')
```

From this we understand, that the image has three color channels. They represent red, green, and blue color intensities. We also understand from the min and max values, that the image is represented by 8 bit unsigned integer, which os also evident from the dtype attribute. We can separate them and visualize.

### RGB channels 

Visualize the levels of each color channel. Pay close attention to the traffic signs!

```python
# Isolate RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

# Visualize the individual color channels
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='gray')
ax2.set_title('G channel')
ax2.imshow(g, cmap='gray')
ax3.set_title('B channel')
ax3.imshow(b, cmap='gray')

## Which area has the lowest value for red? What about for blue?
```

## Key Points to notice

* Traffic Signs and Sky
    * Red Channel
        * For the red channel traffic sign to the left is orange, which is made up of partly red, Hence it is looking brighter due to higher values of the red pixel intensities. Where as at the right traffic for the most parts there is no hint of red. Hence, for the red channel, that part is dark. In lower side of the same sign we have some orange parts, hence that part again looks brighter for the red channel.
        * Sky is indigo, and it has little to no hint of red. Hence sky is the darkest in the red channel.
    * Green Channel
        * Left traffic sign is less bright compared to red channel but it has a gray tone because orange is made of more red and less green parts. Most parts of the right traffic sign is brighter in this channel because it has more green in it.
        * Sky is brighter compared to red channel but not very bright because Sky has less green color in it. It would be brightest in the blue channel.
    * Blue Channel
        * Left traffic sign completely and a part of right traffic sign is dark in blue channel, because there is no blue in it. Leftover part of right traffic signs is less bright than green channel because there is more green in it and less blue.
        * SPecially noticeable for blue channel is the sky. In blue channel sky is the brightest because there is mostly blue in it.

In this way if we analyze the different color channels against the color image we would understand, that the pixel intensities make sense.
