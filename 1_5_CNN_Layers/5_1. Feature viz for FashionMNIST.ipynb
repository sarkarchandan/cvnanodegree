# Visualizing CNN Layers
---
In this notebook, we load a trained CNN (from a solution to FashionMNIST) and implement several feature visualization techniques to see what features this network has learned to extract.

### Load the [data](https://pytorch.org/vision/stable/generated/torchvision.datasets.FashionMNIST.html#fashionmnist)

In this cell, we load in just the **test** dataset from the FashionMNIST class.

```python
# our basic libraries
from typing import List, Iterator
from pathlib import Path
import cv2
import torch

# data loading and transforming
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.transforms import ToTensor
```

```python
# The output of torchvision datasets are PILImage images of range [0, 1]. 
# We transform them to Tensors for input into a CNN

## Define a transform to read the data in as a tensor
data_transform: ToTensor = transforms.ToTensor()

data_dir: str = Path().absolute().joinpath('data').__str__()
test_data: FashionMNIST = FashionMNIST(root=data_dir, 
                        train=False,
                        download=True, 
                        transform=data_transform)

# Print out some stats about the test data
print('Test data, number of images: ', len(test_data))
```

```python
# Prepare data loaders, set the batch_size
## TODO: you can try changing the batch_size to be larger or smaller
## when you get to training your network, see how batch_size affects the loss
batch_size: int = 20

test_loader: DataLoader = DataLoader(dataset=test_data, batch_size=batch_size, shuffle=True)

# specify the image classes
classes: List[str] = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
           'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### Visualize some test data

This cell iterates over the training dataset, loading a random batch of image/label data, using `dataiter.next()`. It then plots the batch of images and labels in a `2 x batch_size/2` grid.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
    
# obtain one batch of training images
dataiter = iter(test_loader)
images, labels = dataiter.next()
images = images.numpy()

# plot the images in the batch, along with the corresponding labels
fig = plt.figure(figsize=(25, 4))
for idx in np.arange(batch_size):
    ax = fig.add_subplot(2, batch_size//2, idx+1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(images[idx]), cmap='gray')
    ax.set_title(classes[labels[idx]])
```

### Define the network architecture

The various layers that make up any neural network are documented, [here](http://pytorch.org/docs/master/nn.html). For a convolutional neural network, we'll use a simple series of layers:
* Convolutional layers
* Maxpooling layers
* Fully-connected (linear) layers

```python
import torch.nn as nn
import torch.nn.functional as F
from torchinfo import summary

class Net(nn.Module):

    conv1: nn.Conv2d
    pool1: nn.MaxPool2d
    conv2: nn.Conv2d
    pool2: nn.MaxPool2d
    fc1: nn.Linear
    fc1_drop: nn.Dropout
    fc2: nn.Linear

    def __init__(self) -> None:
        super(Net, self).__init__()
        
        # 1 input image channel (grayscale), 10 output channels/feature maps
        # 3x3 square convolution kernel
        ## output size = (W-F)/S +1 = (28-3)/1 +1 = 26
        # the output Tensor for one image, will have the dimensions: (10, 26, 26)
        # after one pool layer, this becomes (10, 13, 13)
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=10, kernel_size=3)
        
        # maxpool layer
        # pool with kernel_size=2, stride=2
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # second conv layer: 10 inputs, 20 outputs, 3x3 conv
        ## output size = (W-F)/S +1 = (13-3)/1 +1 = 11
        # the output tensor will have dimensions: (20, 11, 11)
        # after another pool layer this becomes (20, 5, 5); 5.5 is rounded down
        self.conv2 = nn.Conv2d(in_channels=10, out_channels=20, kernel_size=3)

        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # 20 outputs * the 5*5 filtered/pooled map size
        self.fc1 = nn.Linear(in_features=20*5*5, out_features=50)
        
        # dropout with p=0.4
        self.fc1_drop = nn.Dropout(p=0.4)
        
        # finally, create 10 output channels (for the 10 classes)
        self.fc2 = nn.Linear(in_features=50, out_features=10)

    # define the feedforward behavior
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Two conv/relu + pool layers
        x = self.conv1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = F.relu(x)
        x = self.pool2(x)

        # Prep for linear layer
        # this line of code is the equivalent of Flatten in Keras
        x = x.view(x.size(0), -1)
        
        # Two linear layers with dropout in between
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc1_drop(x)
        x = self.fc2(x)
        
        # final output
        return x
```

### Load in our trained net

This notebook needs to know the network architecture, as defined above, and once it knows what the "Net" class looks like, we can instantiate a model and load in an already trained network.

The architecture above is taken from the example solution code, which was trained and saved in the directory `saved_models/`.


```python
# instantiate your Net
net: Net = Net()

# load the net parameters by name
net.load_state_dict(torch.load('saved_models/fashion_net_ex.pt'))

print(f'Network layers: {net}')

print(f'Network summary: {summary(model=net, batch_dim=(20, 1, 28, 28))}')
```

## Feature Visualization

Sometimes, neural networks are thought of as a black box, given some input, they learn to produce some output. CNN's are actually learning to recognize a variety of spatial patterns and you can visualize what each convolutional layer has been trained to recognize by looking at the weights that make up each convolutional kernel and applying those one at a time to a sample image. These techniques are called feature visualization and they are useful for understanding the inner workings of a CNN.

In the cell below, you'll see how to extract and visualize the filter weights for all of the filters in the first convolutional layer.

Note the patterns of light and dark pixels and see if you can tell what a particular filter is detecting. For example, the filter pictured in the example below has dark pixels on either side and light pixels in the middle column, and so it may be detecting vertical edges.

<img src='images/edge_filter_ex.png' width= 30% height=30%/>



### Extract the filters

```python
# Get the weights in the first conv layer. Calling the 
# the data attribute extracts the filters as torch.Tensor type.
# Generally, the weights and biases have the type 
# nn.parameter.Parameter
weights: torch.Tensor = net.conv1.weight.data
w: np.ndarray = weights.numpy()

print(f'Filters array shape: {w.shape}')
```

```python
# For 10 filters
fig=plt.figure(figsize=(20, 8))
columns = 5
rows = 2
for i in range(0, columns*rows):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(w[i][0], cmap='gray')
    
print('First convolutional layer')
plt.show()
```

```python
# Extract and visualize the filters from the second convolution layer
conv2_filters: torch.Tensor = net.conv2.weight.data
wts: np.ndarray = conv2_filters.numpy()
# Visualize the filters
fig: plt.Figure =  plt.figure(figsize=(20, 5))
columns: int = 10
rows: int = 2
for i in range(columns * rows):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(wts[i, 0], cmap='gray')
```

As we have studied in theory, that as the layer goes deeper convolution filters tend to detect more complex patterns in the image. In this example above it is hard to tell if the filter patterns of the second convolution layer is more complex compared to the first convolution layer. That's partly because in this example we are working with pretty low resolution images. Hence our filters are also mostly (3, 3) and it's hard to identify complex patterns or distinguish between the filter patterns of two layers. But now let's move to using these filters on sample images. That might give us a better intuition about, what these filters are actually learning.

### Activation Maps

Next, we'll see how to use OpenCV's `filter2D` function to apply these filters to a sample test image and produce a series of **activation maps** as a result. We'll do this for the first and second convolutional layers and these activation maps whould really give us a sense for what features each filter learns to extract.

```python
# obtain one batch of testing images
dataiter: Iterator = iter(test_loader)
images, labels = dataiter.next()
images: np.ndarray = images.numpy()

# select an image by index
idx: int = 2
img: np.ndarray = np.squeeze(images[idx])

# Use OpenCV's filter2D function 
# apply a specific set of filter weights (like the one's displayed above) to the test image

plt.imshow(img, cmap='gray')
```

#### Apply conv1 layer filters

```python
weights: torch.Tensor = net.conv1.weight.data
w: np.ndarray = weights.numpy()

# 1. first conv layer
# for 10 filters
fig: plt.Figure =plt.figure(figsize=(30, 5))
columns: int = 5*2
rows: int = 2
for i in range(0, columns*rows):
    fig.add_subplot(rows, columns, i+1)
    if ((i%2)==0):
        plt.imshow(w[int(i/2)][0], cmap='gray')
    else:
        c: np.ndarray = cv2.filter2D(src=img, ddepth=-1, kernel=w[int((i-1)/2)][0])
        plt.imshow(c, cmap='gray')
plt.show()
```

#### Apply conv2 layer filters

```python
plt.imshow(img, cmap='gray')
```

```python
# second conv layer, conv2
weights: torch.Tensor = net.conv2.weight.data
w: np.ndarray = weights.numpy()
w.shape
```

```python
# 1. first conv layer
# for 20 filters
fig=plt.figure(figsize=(30, 10))
columns: int = 5*2
rows: int = 2*2
for i in range(0, columns*rows):
    fig.add_subplot(rows, columns, i+1)
    if ((i%2)==0):
        plt.imshow(w[int(i/2)][0], cmap='gray')
    else:
        c: np.ndarray = cv2.filter2D(src=img, ddepth=-1, kernel=w[int((i-1)/2)][0])
        plt.imshow(c, cmap='gray')
plt.show()
```
